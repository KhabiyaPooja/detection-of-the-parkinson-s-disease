# Simplify the data

```python
x = df.iloc[:, 0:754].values
y = df.iloc[:, -1].values 
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
x= StandardScaler().fit_transform(x)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)
```

```python
x_train.shape
```

```python
x_test.shape
```

# PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 6)
x_train_new = pca.fit_transform(x_train)
x_test_new = pca.transform(x_test)   
```

```python
x_train_new.shape
```

```python
x_test_new.shape
```

# LogisticRegression before PCA

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=42)
classifier.fit(x_train,y_train)
```

# LogisticRegression after PCA

```python
classifier2 = LogisticRegression(random_state=42)
classifier2.fit(x_train_new,y_train)
```

```python
y_pred = classifier.predict(x_test) 
y_pred_new = classifier2.predict(x_test_new) 
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)*100
```

```python
accuracy_score(y_test, y_pred_new)*100
```

LogisticRegression after PCA has accuracy upto 86%.
